```python
group1 = df1.groupby('Type 1')['Type 1'].count().sort_values(ascending=False)
group1
```

```python
plt.figure(figsize=(16, 8))
plt.bar(group1.index, group1.values)
plt.title('Number of Type 1')
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.show()
```

```python
group2 = df1.groupby('Type 2')['Type 2'].count().sort_values(ascending=False)
group2
```

```python
plt.figure(figsize=(16, 8))
plt.bar(group2.index, group2.values)
plt.title('Number of Type 2')
plt.xlabel('Category')
plt.ylabel('Quantity')
plt.show()
```

```python
gen = df1.groupby('Generation')['Generation'].count()
gen
```

```python
gen.plot(kind='bar')
plt.title('Number of Pokemon per Generation')
plt.xlabel('Generation')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
legend = df1['Legendary'].value_counts()
per = legend/len(df1) * 100
per
```

```python
labels = ['Not', 'Legendary']
plt.pie(x = legend, labels=labels, autopct= "%1.1f%%");
```

```python
def plotty(vari):
    one = df1.groupby('Type 1')[vari].mean().sort_values(ascending=False)
    plt.figure(figsize=(16, 8))
    plt.bar(one.index, one.values)
    plt.title('Highest ' + vari + ' in Pokemon')
    plt.xlabel('Pokemon Type 1')
    plt.ylabel('Average')
    plt.show()
    
```

```python
plotty('Total')
```
